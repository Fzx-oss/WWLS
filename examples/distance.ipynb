```python
import ot
import sys
import random
import numpy as np
import scipy as sp
import networkx as nx
```

```python
from Distances.WWL import *
from Kernels.utilities import read_labels

def WWL_distance(ig_graphs, h):
    label_sequences = compute_wl_embeddings_discrete(ig_graphs, h)
    print('Computing the Wasserstein distances...')
    wasserstein_distances = compute_wasserstein_distance(label_sequences, h, sinkhorn=False, discrete=True)
    D = wasserstein_distances[h]
    res = D[0] / max(D[0])
    return res
```

```python
sys.path.append("../WWLS")
from tqdm import tqdm
from WWLS import WWLS
from graph_data import Graph

def WWLS_distance(py_graphs, h):
    graphs = [Graph(graph) for graph in py_graphs]
    wwls = WWLS(h)
    graph_features = [wwls.embedding(i, graph) for i, graph in enumerate(tqdm(graphs, desc='Building WL subtrees'))] 
    sparse_graph_features = wwls.to_sparse(graph_features)
    D = wwls.compute_distance_matrices(sparse_graph_features, len(graphs))
    res = D[0] / max(D[0])
    return res
```

```python
from scipy.sparse.csgraph import shortest_path

def GW_distance(nx_graphs):
    graphs = [nx.adjacency_matrix(nx_graph) for nx_graph in nx_graphs]
    sp_matrices = []
    for graph in graphs:
        sp_matrix = shortest_path(graph)
        sp_matrix[sp_matrix == np.inf] = 100
        sp_matrices.append(sp_matrix)
        
    C_list = [sp.spatial.distance.cdist(sp_matrix, sp_matrix) for sp_matrix in sp_matrices]
    
    num_graphs = len(C_list)
    
    C1 = C_list[0]
    p = ot.unif(len(C1))
    res = []
    
    for i in range(num_graphs):
        C2 =  C_list[i]
        q = ot.unif(len(C2))
        gw, log = ot.gromov.gromov_wasserstein(C1, C2, p, q, 'square_loss', verbose=False, log=True)
        res.append(log['gw_dist'])
    
    res = np.array(res)
    res = res / max(res)
    return res
```

```python
from numpy import linalg as LA

def I(n):
    A = []
    for i in range(n):
        A.append([1 if j == i else 0 for j in range(n)])
    return A

def Laplacian(nx_graphs):
    lap_matrices = [nx.laplacian_matrix(nx_graph).toarray() for nx_graph in nx_graphs]
    res1 = []
    lap_matrix1 = lap_matrices[0]
    for lap_matrix2 in lap_matrices:
        X = lap_matrix1 - lap_matrix2
        dis = LA.norm(X,"fro")
        res1.append(dis)
    res1 = np.array(res1)
    res1 = res1 / max(res1)

    num_nodes = len(lap_matrices[0])
    I_matrix = np.array(I(num_nodes))
    P_matrices = []
    permutation_index = list(range(num_nodes))
    for i in range(len(nx_graphs)):
        random.shuffle(permutation_index)
        I_matrix2 = I_matrix[permutation_index]
        P_matrices.append(copy.deepcopy(I_matrix2))
    
    lap_matrices_P = []
    for i, L in enumerate(lap_matrices):
        P = P_matrices[i]
        L2 = P @ L @ P.T
        lap_matrices_P.append(L2)
    
    res2 = []
    lap_matrix_P_1 = lap_matrices[0]
    for lap_matrix_P_2 in lap_matrices_P:
        X = lap_matrix_P_1 - lap_matrix_P_2
        dis = LA.norm(X,"fro")
        res2.append(dis)
    res2 = np.array(res2)
    res2 = res2 / max(res2)

    return res1, res2
```
